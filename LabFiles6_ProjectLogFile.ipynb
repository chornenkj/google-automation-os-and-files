```python
import re
import operator
import csv

# Initializing variables to count data
errors = {}
user_errors = {}
user_infoes = {}

# Iterating through file rows to count data
with open('syslog.log') as file:
    for line in file:
        result_error = re.search(r'ERROR (.*) \((.*)\)',line)
        if result_error:
            errors[result_error.groups()[0]] = errors.get(result_error.groups()[0],0) + 1
            user_errors[result_error.groups()[1]] = user_errors.get(result_error.groups()[1],0) + 1
        result_info = re.search(r'INFO.*\((.*)\)',line)
        if result_info:
            user_infoes[result_info.groups()[0]] = user_infoes.get(result_info.groups()[0],0) + 1

# Sorting elements
errors = sorted(errors.items(), key = operator.itemgetter(1), reverse=True)
user_errors = sorted(user_errors.items())

# Create dictionary with errors
errors_list = []    
for error in errors:
    ed = {}
    ed['Error'] = error[0]
    ed['Count'] = error[1]
    errors_list.append(ed)

# Create dictionary with users
users = []
for user,error in user_errors:
    td = {}
    td['Username'] = user
    if user in user_infoes:
        td['INFO'] = user_infoes[user]
    else:
        td['INFO'] = 0
    td['ERROR'] = error
    users.append(td)

# Write report on users into a file    
keys = ['Username','INFO','ERROR']
with open('user_statistics.csv','w') as file:
    writer = csv.DictWriter(file,fieldnames=keys)
    writer.writeheader()
    writer.writerows(users)
    file.close()

# Write report on errors into a file
keys_e = ['Error','Count']
with open('error_message.csv','w') as file_e:
    writer_e = csv.DictWriter(file_e,fieldnames=keys_e)
    writer_e.writeheader()
    writer_e.writerows(errors_list)
    file.close()
```
